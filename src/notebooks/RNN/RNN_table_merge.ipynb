```python
new_df = pd.read_csv('RNN_data.csv')
new_df
```

```python
new_df = new_df.drop_duplicates()
new_df
```

```python


#adding columns to RNN_DATA
def pad_status(row):
    
    pad_status = 0
    
    # Cascade down by lifecycle event to determine the pad status
    if pd.notnull(row['tst_first_detection_date']):  
        
        pad_status = 1 # Has detection
        
    if pd.notnull(row['tst_permit_date']):

        pad_status = 2 # Has Permit
 
    if pd.notnull(row['tst_spud_date']):

        pad_status = 3 # Has spud
            
    if pd.notnull(row['tst_completion_date']):

        pad_status = 4 # Has one completion confirmation

        if pd.notnull(row['tst_fracfocus_jobstart_date']):

            pad_status = 5 # Has two completion confirmations

            if pd.notnull(row['tst_fracfocus_jobend_date']):

                pad_status = 6 # Has three completion confirmations
        else:

            if pd.notnull(row['tst_fracfocus_jobend_date']):

                pad_status = 5 # Has two completion confirmations
        
            else:
                
                if pd.notnull(row['tst_fracfocus_jobstart_date']):
                    
                    pad_status = 4 # Has one completion confirmation
                
                    if pd.notnull(row['tst_fracfocus_jobend_date']):
                        
                        pad_status = 5 # Has two completion confirmations
                    
                else:
                    
                    if pd.notnull(row['tst_fracfocus_jobend_date']):
                        
                        pad_status = 4 # Has one completion confirmation                     
    
    return pad_status

    
new_df['pad_status'] = new_df.apply(pad_status, axis=1)
```
