```python
from datetime import datetime
import matplotlib.dates as mdates
import pandas as pd
import matplotlib.pyplot as plt
import geopandas
import numpy as np
from matplotlib.pyplot import text
from matplotlib.pyplot import figure

```

```python
# read csv files as df
df1 = pd.read_csv(r'D:\UCL\Term3\tbx-flaring-shihao\data\well_event_data\midland_county_wellevents.csv')

df2 = pd.read_csv(r'D:\UCL\Term3\tbx-flaring-shihao\data\flaring_data\wellpads_with_flaring_midland.csv')

# convert df to GeoDataFrame with longitude and latitude of pads

df_new_1 = geopandas.GeoDataFrame(
    df1, geometry=geopandas.points_from_xy(df1.longitude_pad, df1.latitude_pad))
        
df_new_2 = geopandas.GeoDataFrame(
    df2, geometry=geopandas.points_from_xy(df2.longitude_pad, df2.latitude_pad))
```

```python
# spatial join with the nearst longitude and latitude of pads, based on 44 pads in wellpads_with_flaring_midland.csv
df_new_reverse = geopandas.sjoin_nearest(df_new_1,df_new_2,how = 'right') 
```

```python

# Example of how to parse time series dat
df_nm_flaring = pd.DataFrame()

for idx, row in df_new_reverse.iterrows():
    
    # Convert strings to objects
    ts_mw = eval(row['ts_flaring_power_MW'])
    #ts_gj = eval(row['ts_flaring_energy_daily_GJ'])
    
    # Loop through each of the events
    for idx, t in enumerate(ts_mw):
        
        # Parse time
        tmid = datetime.strptime(t['time'], '%Y-%m-%d %H:%M:%S')
        if t['value'] > 0:
            new_row = {
                # getting all the needed columns with ts_flaring_power_MW
              'wellpads_id':row.wellpads_id,
               'well_id':row.id_left, 
                'tst_first_detection_date':row.tst_first_detection_date,
                'tst_permit_date':row.tst_permit_date,
                'tst_spud_date':row.tst_spud_date,
                'tst_regulator_completion_date':row.tst_regulator_completion_date,
                'tst_fracfocus_jobstart_date':row.tst_fracfocus_jobstart_date,
                'tst_fracfocus_jobend_date':row.tst_fracfocus_jobend_date,
              'ts_flaring_power_MW':ts_mw[idx]['value'],
              #'flaring_power_MW':t['value'],
              'year':int(tmid.year),
              'month':int(tmid.month),
                'day':int(tmid.day)
                
            }
            df_nm_flaring = df_nm_flaring.append(new_row, ignore_index=True)
```

```python
df_nm_flaring
```

```python
# Example of how to parse time series dat
df_nm_flaring2 = pd.DataFrame()

for idx, row in df_new_reverse.iterrows():
    
    # Convert strings to objects
    #ts_mw = eval(row['ts_flaring_power_MW'])
    ts_gj = eval(row['ts_flaring_energy_daily_GJ'])
    
    # Loop through each of the events
    for idx, t in enumerate(ts_gj):
        
        # Parse time
        tmid = datetime.strptime(t['time'], '%Y-%m-%d %H:%M:%S')
        if t['value'] > 0:
            new_row = {
                #  getting all the needed columns with ts_flaring_energy_daily_GJ
              'wellpads_id':row.wellpads_id,
                'well_id':row.id_left,
                'tst_first_detection_date':row.tst_first_detection_date,
                'tst_permit_date':row.tst_permit_date,
                'tst_spud_date':row.tst_spud_date,
                'tst_regulator_completion_date':row.tst_regulator_completion_date,
                'tst_fracfocus_jobstart_date':row.tst_fracfocus_jobstart_date,
                'tst_fracfocus_jobend_date':row.tst_fracfocus_jobend_date,
              'ts_flaring_energy_daily_GJ':ts_gj[idx]['value'],
              #'flaring_power_MW':t['value'],
              'year':int(tmid.year),
              'month':int(tmid.month),
                'day':int(tmid.day)
            }
            df_nm_flaring2 = df_nm_flaring2.append(new_row, ignore_index=True)
```

```python
# inner join the 2 df to get a df with both ts_flaring_power_MW and ts_flaring_energy_daily_GJ

df_nm_flaring_new = pd.merge(df_nm_flaring2,df_nm_flaring, how ='inner',on = ['wellpads_id','year','month','day','tst_first_detection_date',
                                                                              'tst_permit_date','tst_spud_date','tst_regulator_completion_date',
                                                                              'tst_fracfocus_jobstart_date','tst_fracfocus_jobend_date'])
#df_nm_flaring_new = pd.merge(df_nm_flaring2,df_nm_flaring, how ='inner',on = ['wellpads_id','year','month','day','tst_first_detection_date'])
```

```python
# format the datetime by combining year, month, and day
df_nm_flaring_new['Date'] = pd.to_datetime(df_nm_flaring_new[['year','month','day']])
```

```python
# to make sure all the datetime are set as date only, without time 
df_nm_flaring_new['tst_first_detection_date'] = pd.to_datetime(df_nm_flaring_new['tst_first_detection_date']).dt.date
df_nm_flaring_new['tst_permit_date'] = pd.to_datetime(df_nm_flaring_new['tst_permit_date']).dt.date
df_nm_flaring_new['tst_regulator_completion_date'] = pd.to_datetime(df_nm_flaring_new['tst_regulator_completion_date']).dt.date
df_nm_flaring_new['tst_fracfocus_jobstart_date'] = pd.to_datetime(df_nm_flaring_new['tst_fracfocus_jobstart_date']).dt.date
df_nm_flaring_new['tst_fracfocus_jobend_date'] = pd.to_datetime(df_nm_flaring_new['tst_fracfocus_jobend_date']).dt.date
df_nm_flaring_new['tst_spud_date'] = pd.to_datetime(df_nm_flaring_new['tst_spud_date']).dt.date
```

```python
# function to plot 
def plot_flaring_date(group1):
    # passing date as X axis
    x = group1['Date']
    # ts_flaring_energy_daily_GJ as y1, and ts_flaring_power_MW as y2
    y1 = group1['ts_flaring_energy_daily_GJ']
    y2 = group1['ts_flaring_power_MW']
    # setting fig size
    fig, ax1 = plt.subplots(figsize=(15,10))
    
    #putting to Y axis together, then plot the line
    ax2 = ax1.twinx()
    flaring_energy_daily_GJ, =ax1.plot(x, y1 )
    flaring_power_MW,= ax2.plot(x, y2,color='orange')
    
    #set labels 
    ax1.set_xlabel('Date')
    ax1.set_ylabel('GJ',c='#1f77b4')
    ax2.set_ylabel('MW', color='orange')

    # Check for NAT, can't plot with any date is NaN or NaT   
    marker_list_noNAT = []
    # we want the 6 markers below 
    marker_list = np.array([group1['tst_first_detection_date'].iloc[0],group1['tst_permit_date'].iloc[0],group1['tst_spud_date'].iloc[0],
                   group1['tst_regulator_completion_date'].iloc[0],group1['tst_fracfocus_jobstart_date'].iloc[0],
                            group1['tst_fracfocus_jobend_date'].iloc[0]])
    for i in marker_list:
        if str(i) != 'NaT':
            marker_list_noNAT.append(i)
    # check Nat ends

    # getting Y min and Y max to draw vertical lines and annotations
    ymin,ymax = ax2.get_ylim()
    # convert date to num for x axis to show
    marker_list_noNAT = mdates.date2num(marker_list_noNAT)
    # plot makers (vertical line with dot)
    plt.vlines(x=marker_list_noNAT, ymin=ymin, ymax=ymax, linestyle='dotted',color = 'r')

    #ax1.xaxis.set_major_locator(mdates.YearLocator())
    #ax1.xaxis.set_major_formatter(mdates.DateFormatter('%y.%m.%d'))
    
    #text needed for markers
    textlist = ['first_detection_date','permit_date','spud_date','regulator_completion_date',
                'fracfocus_jobstart_date','fracfocus_jobend_date']
    # plot annotation for markers
    for i, x in enumerate(marker_list_noNAT):
        text(x, 0.5, textlist[i], rotation=90, verticalalignment='center') # 0.5 to set text in the middle,rotation to make the text verticl  

    # save fig in default location 
    plt.savefig('flaring_date_wellid_'+ str(group1['well_id_x'].iloc[0])+'.png')
    plt.show()
    return fig
```

```python
df_cal = df_nm_flaring_new.copy()
```

```python
df_cal = df_cal.drop(columns = ['year','month','day','well_id_y'])
```

```python
df_cal
```

```python
df_cal['Date'] = pd.to_datetime(df_cal['Date']).dt.date
df_cal['diff_first_detection_flare'] = (df_cal['Date']-df_cal['tst_first_detection_date'] ).dt.days
df_cal['diff_permit_flare'] = (df_cal['Date'] - df_cal['tst_permit_date']).dt.days
df_cal['diff_spud_flare'] = (df_cal['Date'] - df_cal['tst_spud_date'] ).dt.days
df_cal['diff_regulator_completion_flare'] = ( df_cal['Date'] - df_cal['tst_regulator_completion_date'] ).dt.days
df_cal['diff_fracfocus_jobstart_flare'] = ( df_cal['Date'] - df_cal['tst_fracfocus_jobstart_date'] ).dt.days
df_cal['diff_fracfocus_jobend_flare'] = ( df_cal['Date']- df_cal['tst_fracfocus_jobend_date'] ).dt.days
```

```python
df_cal_result = df_cal.drop(columns = ['tst_first_detection_date','tst_permit_date','tst_spud_date','tst_regulator_completion_date','tst_fracfocus_jobstart_date','tst_fracfocus_jobend_date','Date'])
df_cal_result
```

```python
df_cal_result_agg_by_wellid  = pd.DataFrame()
df_cal_result_agg_by_wellid['avg_diff_first_detection_flare'] =  df_cal_result.groupby(['well_id_x'])['diff_first_detection_flare'].agg('mean')
df_cal_result_agg_by_wellid['avg_diff_permit_flare'] =  df_cal_result.groupby(['well_id_x'])['diff_permit_flare'].agg('mean')
df_cal_result_agg_by_wellid['avg_diff_spud_flare'] =  df_cal_result.groupby(['well_id_x'])['diff_spud_flare'].agg('mean')
df_cal_result_agg_by_wellid['avg_diff_regulator_completion_flare'] =  df_cal_result.groupby(['well_id_x'])['diff_regulator_completion_flare'].agg('mean')
df_cal_result_agg_by_wellid['avg_diff_fracfocus_jobstart_flare'] =  df_cal_result.groupby(['well_id_x'])['diff_fracfocus_jobstart_flare'].agg('mean')
df_cal_result_agg_by_wellid['avg_diff_fracfocus_jobend_flare'] =  df_cal_result.groupby(['well_id_x'])['diff_fracfocus_jobend_flare'].agg('mean')
df_cal_result_agg_by_wellid
```

```python
# plot bar chart and see if there is a relation
avg_diff_first_detection_flare = df_cal_result_agg_by_wellid['avg_diff_first_detection_flare']
avg_diff_permit_flare = df_cal_result_agg_by_wellid['avg_diff_permit_flare']
avg_diff_spud_flare = df_cal_result_agg_by_wellid['avg_diff_spud_flare']
avg_diff_regulator_completion_flare = df_cal_result_agg_by_wellid['avg_diff_regulator_completion_flare']
avg_diff_fracfocus_jobstart_flare = df_cal_result_agg_by_wellid['avg_diff_fracfocus_jobstart_flare']
avg_diff_fracfocus_jobend_flare = df_cal_result_agg_by_wellid['avg_diff_fracfocus_jobend_flare']

 
X = df_cal_result_agg_by_wellid.index

  
X_axis = np.arange(len(X))
  
plt.bar(X_axis-1, avg_diff_first_detection_flare, 0.4, label = 'avg_diff_first_detection_flare')
plt.bar(X_axis-0.6, avg_diff_permit_flare, 0.4, label = 'avg_diff_permit_flare')
plt.bar(X_axis-0.2, avg_diff_spud_flare, 0.4, label = 'avg_diff_spud_flare')
plt.bar(X_axis+0.2, avg_diff_regulator_completion_flare, 0.4, label = 'avg_diff_regulator_completion_flare')
plt.bar(X_axis+0.6, avg_diff_fracfocus_jobstart_flare, 0.4, label = 'avg_diff_fracfocus_jobstart_flare')
plt.bar(X_axis+1, avg_diff_fracfocus_jobend_flare, 0.4, label = 'avg_diff_fracfocus_jobend_flare')
  
plt.xticks(X_axis, X)
plt.xlabel("Groups")
plt.ylabel("Number of Days")
plt.title("Flare vs markers date")
plt.legend()
plt.show()




```
