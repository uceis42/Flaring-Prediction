```python
# trying to use geopandas 
#df_new_1 = geopandas.read_file(r'D:\UCL\Term3\tbx-flaring-shihao\data\well_event_data\midland_county_wellevents.csv')

well_event_gdf = geopandas.GeoDataFrame(
    well_event_df, geometry=geopandas.points_from_xy(well_event_df.longitude_pad, well_event_df.latitude_pad))
        
#df_new_2 = geopandas.read_file(r'D:\UCL\Term3\tbx-flaring-shihao\data\flaring_data\wellpads_with_flaring_midland.csv')   


flaring_pad_gdf = geopandas.GeoDataFrame(
    flaring_pad_df, geometry=geopandas.points_from_xy(flaring_pad_df.longitude_pad, flaring_pad_df.latitude_pad))
```

```python
well_event_gdf = well_event_gdf.drop(columns=['id', 'Unnamed: 0']) # Drop some unwanted columns
well_event_gdf
```

```python
flaring_pad_gdf = flaring_pad_gdf.drop(columns=['id', 'Unnamed: 0']) # Drop some unwanted columns
flaring_pad_gdf
```

```python
well_event_gdf = well_event_gdf.set_crs(4326).to_crs(3857) # Convert to projected CRS (EPSG:3857) from geographic
flaring_pad_gdf = flaring_pad_gdf.set_crs(4326).to_crs(3857)
```
