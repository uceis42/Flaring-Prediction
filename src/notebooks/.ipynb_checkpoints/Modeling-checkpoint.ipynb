```python
from datetime import datetime
import matplotlib.dates as mdates
import pandas as pd
import matplotlib.pyplot as plt
import geopandas
import seaborn as sns
import numpy as np
from matplotlib.pyplot import text
from matplotlib.pyplot import figure
import os
import glob
```

```python
path = r'D:\UCL\Term3\tbx-flaring-shihao\data\well_event_to_flaring_v1\data\model_data' 
csv_files = glob.glob(os.path.join(path, "*.csv"))
```

```python

csv_files
```

```python
#merge 2 csv files (in one group) into one csv 
for index in range(186):  
    df_events = pd.read_csv(csv_files[2*index])
    df_flaring = pd.read_csv(csv_files[2*index+1])
    df_concat = pd.concat([df_events, df_flaring], axis=1)
    if pd.isnull(df_concat.iloc[-1]['api']):
        df_concat= df_concat.ffill()
    df_concat.to_csv(str(index)+'.csv')
        
        
```

```python
# merge each group's csv to one file
path = r'D:\UCL\Term3\tbx-flaring-shihao\src\notebooks' 
csv_files = glob.glob(os.path.join(path, "*.csv"))
li = []
for file in csv_files:
    df = pd.read_csv(file, index_col=None, header=0)
    li.append(df)
df = pd.concat(li, axis=0, ignore_index=True)    
df.to_csv('final_data.csv')     
```
