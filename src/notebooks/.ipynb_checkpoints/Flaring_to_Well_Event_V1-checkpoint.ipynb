```python
from datetime import datetime
import matplotlib.dates as mdates
import pandas as pd
import matplotlib.pyplot as plt
import geopandas
import seaborn as sns
import numpy as np
from matplotlib.pyplot import text
from matplotlib.pyplot import figure
import os
from shapely.geometry import Point, MultiPoint
from shapely.ops import nearest_points

from esda.adbscan import ADBSCAN, get_cluster_boundary, remap_lbls
```

```python
# read wellpad csv data as df
well_reg_df = pd.read_csv(os.path.join('data','wellpad_data', 'wells_regulator_permian.csv'))
# read gpkg as geodf
flaring_event_df = geopandas.read_file(os.path.join('data','flaring_data', 'viirs_permian_FRP_ALL.gpkg'))
flaring_event_df = flaring_event_df.to_crs("epsg:3857")

# convert regulator event data to geodataframe
well_reg_df = geopandas.GeoDataFrame(well_reg_df,
                                 geometry=geopandas.points_from_xy(well_reg_df.long, well_reg_df.lat), crs='epsg:4326')
well_reg_df_3857 = well_reg_df.to_crs("epsg:3857")
well_reg_df = well_reg_df_3857.set_crs('epsg:3857')


# Load wellpad data and attributes
wellpad_df = pd.read_csv(os.path.join('data','wellpad_data', 'permian_wellpads.csv'))
wellpad_df = geopandas.GeoDataFrame(wellpad_df, geometry=geopandas.points_from_xy(wellpad_df.longitude_pad, wellpad_df.latitude_pad), crs='epsg:4326')
wellpad_df = wellpad_df.to_crs("epsg:3857")
wellpad_df = wellpad_df.rename(columns={'pad_id':'wellpad_code'})

# Load estimate of number of wells (wellheads) for each pad
wells_per_wellpad = pd.read_csv(os.path.join('data','wellpad_data', 'wells_per_wellpad.csv'))
wells_per_wellpad = wells_per_wellpad[['wellpads_id', 'wellpad_code', 'wells']]


# Merge estimated wellheads with wellpad dataframe
wellpad_df = pd.merge(wellpad_df, wells_per_wellpad, how='left', on='wellpad_code')
             

## Load lease level data
leases_poly =  geopandas.read_file(os.path.join('data','operator_data', 'leases_og.gpkg'))


well_pad_df = well_reg_df.merge(wellpad_df[['wellpad_code', 'wellpads_id', 'wells',
                                            'area_m2', 'tst_first_detection_date', 
                                            'company_type']] , on='wellpad_code')


```

```python
# Cluster flaring over centroids

flaring_event_df["X"] = flaring_event_df.geometry.x
flaring_event_df["Y"] = flaring_event_df.geometry.y
flaring_event_df

adbs = ADBSCAN(375, 3, pct_exact=0.5, reps=10, keep_solus=True)
adbs.fit(flaring_event_df)
```

```python
# Change lbls attribute to "spatial group"
flaring_event_df = flaring_event_df.assign(spatial_group=adbs.votes["lbls"])
flaring_event_df['spatial_group'] = pd.to_numeric(flaring_event_df['spatial_group'])
flaring_event_df = flaring_event_df.loc[flaring_event_df['spatial_group'] > 0]
```

```python
flaring_event_df.to_file('flaring_clusters.geojson') # Save clusters to geojson for visualisation
```

```python
# Temporal clustering

import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib import cm
from matplotlib.dates import DateFormatter
import seaborn as sns
import pandas as pd

# Handle date time conversions between pandas and matplotlib
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

# Use white grid plot background from seaborn
sns.set(font_scale=1.5, style="whitegrid")

tab_cmap = ['#32a852',
    '#5c62b5',
    '#e587e8',
    '#bdb542',
    '#38a69e',
    '#444af2',
    '#a80274',
    '#a86302',
    '#367fff',
    '#367fff']

tab_cmap = tab_cmap + tab_cmap + tab_cmap + tab_cmap # Colourmap

flaring_event_clusters = flaring_event_df.groupby('spatial_group')

flaring_and_temportal_event_clusters = pd.DataFrame()

cluster_inc = 0

for g in flaring_event_clusters:
    
    time_cluster_id = 0 # tci
    
    
    df = g[1]
    
    df['aq_date'] = pd.to_datetime(df['aq_date'])
    
    df = df.sort_values('aq_date') # Sort by aquisition date of detections
    df['t_diff'] = df["aq_date"].diff() # Discrete time between detections

    tcid = []
    colours = []
    days_diffs = []
    row_inc = 1
    
    # Iterate through each flaring detection in a cluster
    for k,r in df.iterrows():
        
        days_diff = r['t_diff'] /np.timedelta64(1, 'D')
        
        if np.isnan(days_diff):
            days_diff = 0
        else:
            days_diff = int(days_diff)
        
        if days_diff > 50: # If next flare is 50 days after last one, give new tcid
            time_cluster_id+=1

        tcid.append(time_cluster_id)
        colours.append(tab_cmap[time_cluster_id])
        row_inc+=1
        
        days_diffs.append(days_diff)
                
    df['days_diffs'] = days_diffs
    df['temporal_group'] = tcid
    df['color'] = colours
    
    # plot example
    if cluster_inc == 4:
        fig, ax = plt.subplots(1,1,figsize=(20,10))
        df.plot.scatter(x='aq_date', y='power', ax=ax, color=df['color'], s=100)
        date_form = DateFormatter("%m-%Y")
        ax.xaxis.set_major_formatter(date_form)
        plt.xticks(rotation=45, ha='right')
        plt.show()
    
    cluster_inc+=1
    
    flaring_and_temportal_event_clusters = pd.concat([flaring_and_temportal_event_clusters, df])
```

```python
# Let's create new centroids for every temporal and spatially grouped clusters

# spatial group ID = spatial_group
# temporal group ID = temporal_group

# Group by DBSCAN cluster (spatial_group) and temporal cluster id (temporal_group)
flaring_event_groups = flaring_and_temportal_event_clusters.groupby(['spatial_group', 'temporal_group'])

# Find centroid of each cluster
flaring_event_groups_centroids = flaring_event_groups['X', 'Y'].agg('mean').reset_index()

# Make GeoDataFrame
flaring_event_groups_centroids_gdf = geopandas.GeoDataFrame(flaring_event_groups_centroids,
            geometry=geopandas.points_from_xy(flaring_event_groups_centroids.X, flaring_event_groups_centroids.Y), crs='epsg:3857')

# Find intersecting lease polygons
flaring_event_groups_centroids_with_lease = geopandas.sjoin(flaring_event_groups_centroids_gdf,leases_poly,how='left',op='within')

```

```python
def plot_flaring_date(flare_episode, well_event, fname):
    
    # passing date as X axis
    x = flare_episode.aq_date
    # ts_flaring_energy_daily_GJ as y1, and ts_flaring_power_MW as y2
    y1 = flare_episode.power

    # setting fig size
    fig, ax1 = plt.subplots(figsize=(15,10))
    
    #putting to Y axis together, then plot the line
    ax2 = ax1.twinx()
#     flaring_energy_daily_GJ, =ax1.plot(x, y1)
    flaring_power_MW,= ax1.plot(x, y1,color='orange', marker='*')
    
    #set labels 
    ax1.set_xlabel('Date')
    ax1.set_ylabel('GJ',c='#1f77b4')
#     ax2.set_ylabel('MW', color='orange')

    # Check for NAT, can't plot with any date is NaN or NaT   
    marker_list_noNAT = []
    # we want the 6 markers below 
    marker_list = np.array([well_event['tst_first_detection_date'],well_event['tst_permit_date'],well_event['tst_spud_date'],
                   well_event['tst_completion_date'],well_event['tst_fracfocus_jobstart_date'],
                            well_event['tst_fracfocus_jobend_date']])
    for i in marker_list:
        if str(i) != 'NaT':
            marker_list_noNAT.append(i)
    # check Nat ends

    # getting Y min and Y max to draw vertical lines and annotations
    ymin,ymax = ax2.get_ylim()
    # convert date to num for x axis to show
    marker_list_noNAT = mdates.date2num(marker_list_noNAT)
    # plot makers (vertical line with dot)
    plt.vlines(x=marker_list_noNAT, ymin=ymin, ymax=ymax, linestyle='dotted',color = 'r')

    #ax1.xaxis.set_major_locator(mdates.YearLocator())
    #ax1.xaxis.set_major_formatter(mdates.DateFormatter('%y.%m.%d'))
    
    #text needed for markers
    textlist = ['first_detection_date','permit_date','spud_date','tst_completion_date',
                'fracfocus_jobstart_date','fracfocus_jobend_date']
    # plot annotation for markers
    for i, x in enumerate(marker_list_noNAT):
        text(x, 0.5, textlist[i], rotation=90, verticalalignment='center') # 0.5 to set text in the middle,rotation to make the text verticl  

    # save fig in default location 
    plt.title(well_event.wellpad_code+' '+str(well_event.api)+' '+fname)
    fig.autofmt_xdate()

    plt.savefig(os.path.join('results/figures',well_event.wellpad_code+' '+str(well_event.api)+' '+fname+'.png'))
#     plt.show()
    plt.close()
    return fig
```

```python
# Find regulator events that coincide with both regulator events

# THIS WILL DELETE AND REPLACE CONTENTS OF ./results !!

import warnings
warnings.filterwarnings('ignore')

from shapely.geometry import Point, LineString
import shutil

lease_geoms = []
lease_wellpads = []

def completion_day_diff(row):
    
    completion_day_diffs = []

    if row['tst_completion_date'] is not pd.NaT:
        completion_day_diffs.append((row['flare_t_start'] - row['tst_completion_date']).days)

    if row['tst_fracfocus_jobstart_date'] is not pd.NaT:
        completion_day_diffs.append((row['flare_t_start'] - row['tst_fracfocus_jobstart_date']).days)
        
    if row['tst_fracfocus_jobend_date'] is not pd.NaT:
        completion_day_diffs.append((row['flare_t_start'] - row['tst_fracfocus_jobend_date']).days)
    
    if len(completion_day_diffs):
        
        # Find minimum absolute value
        absolute = np.abs(completion_day_diffs)
        idx = np.where(absolute == np.min(absolute))[0][0]
        return completion_day_diffs[idx]

    else:
        return np.nan
    
    return np.min(np.abs(completion_day_diffs))

def spud_day_diff(row):
    if row['tst_spud_date'] is not pd.NaT:
        return (row['flare_t_start'] - row['tst_spud_date']).days
    else:
        return np.nan
    
def permit_day_diff(row):
    if row['tst_permit_date'] is not pd.NaT:
        return (row['flare_t_start'] - row['tst_permit_date']).days
    else:
        return np.nan
        
def detection_day_diff(row):
    if row['tst_first_detection_date'] is not pd.NaT:
        return (row['flare_t_start'] - row['tst_first_detection_date']).days
    else:
        return np.nan

def best_estimate_diff(row):
    
    # Cascade down by lifecycle event to find the best data available
        
    if np.isnan(row['comp_diff']):
        
        if np.isnan(row['spud_diff']):
            
            if np.isnan(row['permit_diff']):
                
                if np.isnan(row['detection_diff']):

                    return row['detection_diff']
                else:
                    return np.nan
                
            else:
                return row['permit_diff']
            
        else:
            return row['spud_diff']
        
    else:
        return row['comp_diff']
    
shutil.rmtree('results')
os.makedirs('results/data', exist_ok=True)
os.makedirs('results/figures', exist_ok=True)


well_event_pads_with_flaring = pd.DataFrame() # Wellpad IDs with flaring matches

for idx, row in flaring_event_groups_centroids_with_lease.iterrows():
    
    # Find lease geometry so we only select candidate wellpads from within the current lease
    
    lease_geom = leases_poly.loc[leases_poly.index==row.index_right].geometry
    
    if not lease_geom.empty: # If lease geometry exists
        
        lease_geoms.append(lease_geom.values[0])
        
        wellpad_centroids_within = well_pad_df.loc[well_pad_df.within(lease_geom.values[0])]
        
        wellpad_centroids_within = wellpad_centroids_within.drop_duplicates()
        
        if len(wellpad_centroids_within) > 0:
            
            wellpad_centroids_within['tst_first_detection_date'] = pd.to_datetime(wellpad_centroids_within['tst_first_detection_date'])
            wellpad_centroids_within['tst_permit_date'] = pd.to_datetime(wellpad_centroids_within['tst_permit_date'])
            wellpad_centroids_within['tst_completion_date'] = pd.to_datetime(wellpad_centroids_within['tst_completion_date'])
            wellpad_centroids_within['tst_fracfocus_jobstart_date'] = pd.to_datetime(wellpad_centroids_within['tst_fracfocus_jobstart_date'])
            wellpad_centroids_within['tst_fracfocus_jobend_date'] = pd.to_datetime(wellpad_centroids_within['tst_fracfocus_jobend_date'])
            wellpad_centroids_within['tst_spud_date'] = pd.to_datetime(wellpad_centroids_within['tst_spud_date'])
            wellpad_centroids_within['tst_permit_date'] = pd.to_datetime(wellpad_centroids_within['tst_permit_date'])
            wellpad_centroids_within['wellpad_vintage_year']=  pd.to_datetime(wellpad_centroids_within['tst_spud_date']).dt.year


            # Load flare spatial-temporal cluster
            flare_episode = flaring_event_groups.get_group((row.spatial_group, row.temporal_group))
            flare_t_start = flare_episode['aq_date'].min()
            flare_t_end = flare_episode['aq_date'].max()

            wellpad_centroids_within['flare_t_start'] = flare_t_start
            wellpad_centroids_within['flare_t_end'] = flare_t_end

            wellpad_centroids_within['comp_diff'] = np.abs(wellpad_centroids_within.apply(completion_day_diff, axis=1))
            wellpad_centroids_within['spud_diff'] = np.abs(wellpad_centroids_within.apply(spud_day_diff, axis=1))
            wellpad_centroids_within['permit_diff'] = np.abs(wellpad_centroids_within.apply(permit_day_diff, axis=1))
            wellpad_centroids_within['detection_diff'] = np.abs(wellpad_centroids_within.apply(detection_day_diff, axis=1))

            wellpad_centroids_within['best_estimate_diff'] = wellpad_centroids_within.apply(best_estimate_diff, axis=1)

            wellpad_centroids_within.sort_values('best_estimate_diff', inplace=True)



            smallest_diff = wellpad_centroids_within.head(1)['best_estimate_diff'].values[0]

            if smallest_diff < 100: # Cutoff is 100 days
                
                fname =str(row.spatial_group)+'_'+str(row.temporal_group)
                
                save_path_folder = os.path.join('results/data',fname)
                os.makedirs(os.path.join('results/data',fname), exist_ok=True)
                
                well_event_matches = wellpad_centroids_within.loc[wellpad_centroids_within['best_estimate_diff']
                                                                  ==smallest_diff]
                
                well_event_matches.drop_duplicates(inplace=True)
                
                well_event_matches.to_csv(os.path.join(save_path_folder,fname+'_events.csv'))
                
                well_event_pads_with_flaring = pd.concat([well_event_pads_with_flaring, well_event_matches])
                
                flare_episode.to_csv(os.path.join(save_path_folder,fname+'_flaring.csv'))
                
                for idx, row in well_event_matches.iterrows():
                    plot_flaring_date(flare_episode, row, fname)  
                
            else:
                well_event_matches = []
```

```python
# Find and tabulate well events + pads with and without flaring


well_pad_df_no_flaring = well_pad_df.loc[~well_pad_df['wellpads_id'].isin(well_event_pads_with_flaring.wellpads_id.unique())]

# Add missing columns
well_pad_df_no_flaring['tst_first_detection_date'] = pd.to_datetime(well_pad_df_no_flaring['tst_first_detection_date'])
well_pad_df_no_flaring['tst_permit_date'] = pd.to_datetime(well_pad_df_no_flaring['tst_permit_date'])
well_pad_df_no_flaring['tst_completion_date'] = pd.to_datetime(well_pad_df_no_flaring['tst_completion_date'])
well_pad_df_no_flaring['tst_fracfocus_jobstart_date'] = pd.to_datetime(well_pad_df_no_flaring['tst_fracfocus_jobstart_date'])
well_pad_df_no_flaring['tst_fracfocus_jobend_date'] = pd.to_datetime(well_pad_df_no_flaring['tst_fracfocus_jobend_date'])
well_pad_df_no_flaring['tst_spud_date'] = pd.to_datetime(well_pad_df_no_flaring['tst_spud_date'])
well_pad_df_no_flaring['tst_permit_date'] = pd.to_datetime(well_pad_df_no_flaring['tst_permit_date'])
well_pad_df_no_flaring['wellpad_vintage_year']=  pd.to_datetime(well_pad_df_no_flaring['tst_spud_date']).dt.year
well_pad_df_no_flaring['flare_t_start'] = None
well_pad_df_no_flaring['flare_t_end'] = None
well_pad_df_no_flaring['comp_diff'] = None    
well_pad_df_no_flaring['spud_diff'] = None       
well_pad_df_no_flaring['permit_diff'] = None        
well_pad_df_no_flaring['detection_diff'] = None  
well_pad_df_no_flaring['best_estimate_diff'] = None


def pad_status(row):
    
    pad_status = 0
    
    # Cascade down by lifecycle event to determine the pad status
    if row['tst_first_detection_date'] is not pd.NaT:
        
        pad_status = 1 # Has detection
        
    if row['tst_permit_date'] is not pd.NaT:

        pad_status = 2 # Has Permit
 
    if row['tst_spud_date'] is not pd.NaT:

        pad_status = 3 # Has spud
            
    if row['tst_completion_date'] is not pd.NaT:

        pad_status = 4 # Has one completion confirmation

        if row['tst_fracfocus_jobstart_date'] is not pd.NaT:

            pad_status = 5 # Has two completion confirmations

            if row['tst_fracfocus_jobend_date'] is not pd.NaT:

                pad_status = 6 # Has three completion confirmations
        else:

            if row['tst_fracfocus_jobend_date'] is not pd.NaT:

                pad_status = 5 # Has two completion confirmations
        
            else:
                
                if row['tst_fracfocus_jobstart_date'] is not pd.NaT:
                    
                    pad_status = 4 # Has one completion confirmation
                
                    if row['tst_fracfocus_jobend_date'] is not pd.NaT:
                        
                        pad_status = 5 # Has two completion confirmations
                    
                else:
                    
                    if row['tst_fracfocus_jobend_date'] is not pd.NaT:
                        
                        pad_status = 4 # Has one completion confirmation                     
    
    return pad_status

    
all_wellpad_events = pd.concat([well_event_pads_with_flaring, well_pad_df_no_flaring])
all_wellpad_events['pad_status'] = all_wellpad_events.apply(pad_status, axis=1)
all_wellpad_events.to_csv(os.path.join('results/data','all_wellpad_events.csv'))

```
